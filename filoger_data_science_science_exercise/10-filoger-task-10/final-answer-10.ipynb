    <p style="text-align: center; font-size: 40px; color:white; font-weight:bolder; line-height:50px; text-transform: 				capitalize;">
            We'll now examine if failure to allocate the requested room type contributes to booking cancellations.
	</p>

```python
def check_different_room(x):
    if x['reserved_room_type'] != x['assigned_room_type']:
        return 1
    else:
        return 0

h_book['different_room_assigned'] = h_book.apply(lambda x: check_different_room(x), axis=1)
cancellation_groupby = h_book.groupby('is_canceled')

room_assignment_df = pd.DataFrame((cancellation_groupby['different_room_assigned'].sum() / cancellation_groupby.size()) * 100).rename(columns={0: 'percentage_different_room_assigned'})

plt.figure(figsize=(10, 7))
sns.barplot(x=room_assignment_df.index, y=room_assignment_df['percentage_different_room_assigned'])
plt.title('Percentage of Bookings with Different Assigned Room Types')
plt.xlabel('Booking Cancellation Status')
plt.ylabel('% Bookings with Different Assigned Rooms')
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])  # Custom x-axis labels
plt.show()
```
