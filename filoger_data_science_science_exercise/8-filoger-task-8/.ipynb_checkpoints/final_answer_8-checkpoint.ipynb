<div style="background-color: #adf480; border-radius: 25px; display: flex; justify-content: center; align-items: center; flex-direction: column;">
        <p style="text-align: left; font-size: 20px; color:black; font-weight:bolder; padding:0px 20px">
        ID : A unique identifier for each flight record in the dataset.<br>
        YEAR : The year in which the flight took place (2013 in this dataset).<br>
        MONTH : The month in which the flight took place (1 to 12).<br>
        DAY : The day of the month on which the flight took place (1 to 31).<br>
        DEP_TIME: The actual local departure time of the flight, in 24-hour format (hhmm).<br> 
        SCHED_DEP_TIME : The scheduled local departure time of the flight, in 24-hour format (hhmm).<br>
        DEP_DELAY : The difference between the actual and scheduled departure times of the flight, in minutes. A positive value indicates a delayed departure, while a negative value indicates an early departure.<br>
        ARR_TIME : The actual local arrival time of the flight, in 24-hour format (hhmm).<br>
        SCHED_ARR_TIME : The scheduled local arrival time of the flight, in 24-hour format (hhmm).<br>
        ARR_DELAY : The difference between the actual and scheduled arrival times of the flight, in minutes. A positive value indicates a delayed arrival, while a negative value indicates an early arrival.<br>
        CARRIER : The two-letter code of the airline carrier for the flight.<br>
        FLIGHT : The flight number of the flight.<br>
        TAILNUM : The unique identifier of the aircraft used for the flight.<br>
        ORIGIN : The three-letter code of the airport of origin for the flight.<br>
        DEST : The three-letter code of the destination airport for the flight.<br>
        AIT_TIME : The duration of the flight, in minutes.<br>
        DISTANCE : The distance between the origin and destination airports, in miles.<br>
        HOUR : The hour component of the scheduled departure time, in local time.<br>
        MINUTE : The minute component of the scheduled departure time, in local time.<br>
        TIME_HOUR : The scheduled departure time of the flight, in local time and format (yyyy-mm-dd hh:mm:ss).<br>
        NAME : The name of the airline carrier for the flight<br>
        </p>
    </div>

<p style="background-color: #adf480;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px; color:black;">Distribution chart of numerical column values<br>
Histogram graphs of numerical values</p>

```python
numerical_columns = ['dep_delay' , 'arr_delay' , 'air_time' , 'distance']

for column in numerical_columns :
    plt.figure(figsize=(20,10))
    sns.histplot(data=df , x=column , bins=200)
    plt.xlabel(column)
    plt.ylabel('Counts')
    plt.title(f'Distribution of {column}')
```

<p style="background-color: #adf480;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px; color:black;">Display a box plot to check the distribution and dispersion of numerical values:</p>

```python
for column in numerical_columns :
    plt.figure(figsize=(20,10))
    sns.boxplot(data=df , x=column , showfliers=False)
    plt.xlabel(column)
    plt.ylabel('Counts')
    plt.title(f'box of {column}')
```

<p style="background-color: #adf480;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px; color:black;">Analysis of relationships between variables <br>
Display the correlation diagram (heatmap) between numerical variables:</p>

```python
plt.figure(figsize=(20,10))

corr = df[numerical_columns].corr()

sns.heatmap(corr , annot=True)

plt.title('correlation between numerical columns')

plt.show()
```

<p style="background-color: #adf480;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px; color:black;">A diagram of the relationship between numerical variables</p>

<p style="background-color: #adf480;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px; color:black;">Analysis of time patterns and trends <br>
Analysis of the distribution of the number of flights by month</p>

```python
df['month'] = pd.to_datetime(df['month'], format='%m')
monthley_flights = df['month'].value_counts().sort_index()

plt.figure(figsize=(10,6))
monthley_flights.plot(marker='o');
plt.xlabel('Month')
plt.ylabel('Number of flights')
plt.title('Distribution of the number of flights by month')
plt.grid(True)
plt.show()
```

<p style="background-color: #adf480;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px; color:black;">Analysis of the distribution of the number of flights by year</p>

```python
df['year'] = pd.to_datetime(df['year'], format='%Y')
monthley_flights = df['year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
monthley_flights.plot(marker='o');
plt.xlabel('year')
plt.ylabel('Number of flights')
plt.title('Distribution of the number of flights by year')
plt.grid(True)
plt.show()
```

```python
def missing_percent (data):
    nan_percent = (data.isnull().sum()/len(data)) * 100
    nan_percent = nan_percent[nan_percent > 0].sort_values()
    return nan_percent
```

```python
nan_percent = missing_percent(data)
nan_percent
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=nan_percent.index , y=nan_percent)
plt.xticks(rotation=90);
```
